```python
import furniture_bench


from src.gym import get_env

from tqdm import trange, tqdm
import matplotlib.pyplot as plt

import os
```

```python
env = get_env(
    obs_type="image",
    num_envs=4,
    gpu_id=1,
)
```

```python
def show_obs(obs):
    fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(16, 8))

    for i in range(4):
        ax[0, i].imshow(obs["color_image1"][i].cpu())
        ax[0, i].axis("off")
        ax[0, i].set_title(f"Env {i}, color_image1")

        ax[1, i].imshow(obs["color_image2"][i].cpu())
        ax[1, i].axis("off")
        ax[1, i].set_title(f"Env {i}, color_image2")

    plt.show()
```

```python
obs = env.reset()

# Show the first observation
show_obs(obs)
```

```python
# Perform 100 random actions
for _ in trange(100):
    obs, reward, done, _ = env.step(env.action_space.sample())

# Show the observation after 100 random actions
show_obs(obs)
```

```python
# Reset and show the observation again
obs = env.reset()
show_obs(obs)
```

```python
env.close()
```
